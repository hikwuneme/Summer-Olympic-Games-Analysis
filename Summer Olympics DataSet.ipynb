# Import Libraries

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Scraping Data On Host Cities and Reading to a Pandas Dataframe

# Scraping Data On List of Participating Countries

```python
url = 'https://en.wikipedia.org/wiki/List_of_participating_nations_at_the_Summer_Olympic_Games'
table_list = pd.read_html(url)
```

```python
df_countries = table_list[2]
df_countries
```

# Cleaning and Manipulating The Data To Get Required Information

```python
#Keeping only the columns needed
df_countries = df_countries[['A', 'Total']]
df_countries
```

```python
#Droping rows with NaN
df_countries = df_countries.dropna()
df_countries
```

```python
#Removing rows where total is the string: total instead of a numeric value
df_countries = df_countries[df_countries.Total != 'Total']
df_countries
```

```python
#Updating the column names
column1 = ['Name', 'Olympic Appearances']
```

```python
df_countries.columns = column1
df_countries
```

# Analysing The Data Gotten to Answer Questions

# 1. Which country(ies) has/have hosted the Olympic Games the most?

```python
df_host.groupby('Country').count()
```

```python
plt.figure(figsize = (25,10))
sns.countplot(x = 'Country', data = df_host)
```

# We see that the United States has hosted the games 5 times, followed by the United Kingdom who has hosted the games 4 times

# 2. Which continent has hosted the games the most?

```python
df_host.groupby('Continent').count()
```

```python
plt.figure(figsize = (25,10))
sns.countplot(x = 'Continent', data = df_host)
```

# Europe has hosted the Olympics a total of 20 times, South America has hosted once and Africa has never hosted the Olympic games

# 3. What has been the male to female representation at the Olympics

```python
df_athletes['Year'] = df_athletes['Year'].astype(str)
df_athletes['Men'] = df_athletes['Men'].astype(int)
df_athletes['Women'] = df_athletes['Women'].astype(int)
```

```python
df_melted = df_athletes[['Year', 'Men', 'Women']]
```

```python
df_melted = pd.melt(df_melted, 'Year', var_name ='Men & Women', value_name = 'Number')
```

```python
sns.set_style('darkgrid')
plt.figure(figsize = (25,10))
sns.lineplot(data=df_melted, x="Year", y="Number", hue="Men & Women", marker ='o')
```

# We see that over the years, there has been a wide gap in male and female represenation at the Olympics, however in recent times, that gap has been closing up with the recent Tokyo 2020 games having almost equal male and female participation

# 4. Which countries have participated in the Olympics the most?

```python
df_countries.info()
```

```python
#First Change Olympic Appearances to type int
df_countries['Olympic Appearances'] = df_countries['Olympic Appearances'].astype(int)
```

```python
#Sort by Olympic Appearances
df_countries_new = df_countries.sort_values(by = ['Olympic Appearances'], ascending = False)
df_countries_new
```

```python
#View the top 10 participating countries
df_countries_new[1:11]
```

# We see that 4 countries(France, Switzerland, Great Britain and Greece) have participated in all 29 Olympic Games while United States, Austria, Denmark, Sweden and Italy have participated in all but one
